## __Option Basics__


#### __Options Lexicon and Notation:__

<br>

__Option__ 
  
__Right__, but not the obligation, to buy or sell a specified quantity of the underlying asset at a
specified price on or before a specified date.
  
__Underlying__ $S$
  
The asset, which the option buyer has the right to buy or sell. Notation: $S$ or $S_t = S(t)$
    
__Call__ $C$ 
  

The right to __BUY__ a specified quantity of the specified underlying asset.    

__Put__ $P$ 
 
The right to __SELL__ a specified quantity of the specified underlying asset.

__Strike/Exercise Price__ $K$ 

The price at which the underlying asset can be bought or sold. 

__Option Expiration__ $T$ 

The date when the option expires.

__Time to Expiration__ 

$τ = T - t$, where $t$ is today's date and $T$ is the option expiration 
 

__Option Types__

|  <font size=2>__Type__</font> |  <font size=3>__The right to exercise__</font> |
| :---     | ---------: |
| <font size=2>__European__ </font>| <font size=2> only at expiration </font> |
| <font size=2>__American__</font>| <font size=2> at any time prior to expiration </font> |
| <font size=2>__Bermudan__</font>| <font size=2> at expiration, as well as some additional dates </font> |


__Physical Settlement__ 

When option exercise results in the purchase or sale of the actual underlying asset.

__Cash Settlement__ 

Results in the transfer of cash to the option holder amounting to the difference between
the strike price and market price of the underlying asset at the time of the option exercise.

__Premium__

The amount paid by the option buyer to the seller (writer). 

__Intrinsic Value at time__ $t$

__*Call*__: $S_t - K$ or $0$, whichever is greater.<br/>
__*Put*__: $K - S_t$ or $0$, whichever is greater.

|  <font size=3>__Moneyness__</font> |  <font size=3>__Intrinsic value__</font> |
| :---     | ---------: |
| <font size=2>In-the-money (__ITM__)  </font>| <font size=2>  Positive </font> |
| <font size=2>Out-of-the-money (__OTM__) </font>| <font size=2>  Zero </font> |
| <font size=2>At-the-money (__ATM__)  </font>| <font size=2>  $S_t = K$  </font> |


__Risk-Free Rate__ $r$

Risk-free rate (continuously compounded; assume that the term structure of interest rates is flat)
 
__Cash Dividends__ $I_0$

PV of all cash income (dividends) to be paid by the underlying asset during the optionís lifetime

__Dividend Yield__ $q$

Dividend yield (with continuous compounding)

__Volatility__ $σ$

Volatility of the underlying asset price movements

#### __Put-Call-Forward Parity__


The payoff of the long forward position at maturity $T$ can be presented as  

$V(T,T) = S_T - K = (S_T - K)^{+} - (K-S_T )^{+}$

Let $C$ and $P$ be the call and put prices at time zero. Then 

$ C - P = V(0,T) = e^{-rT}(F(0,T)-K)$

3 cases:

1. Underlying pays no dividends

  $ C - P = S -  e^{-rT}K$
  
2. Known cash dividends

 $ C - P = S - I_0 - e^{-rT}K$

3. Known dividend yield 
 
  $ C - P = e^{-qT}S -  e^{-rT}K$

#### __Lower and Upper Bounds for European Option Prices :__

Consider the case, when underlying pays no dividends

1. __Call option__ <br/>
   $(S-e^{-rT}K)^{+}\le C \le S$

2. __Put option__<br/>
    $(e^{-rT}K - S)^{+}\le P \le e^{-rT}K$

Same argument applies for the cases with known cash dividends or dividend yield
because the put-call-forward parity still holds (just use the appropriate expression for the
present value of the forward contract $V$). 
   
  
<br>  

<a id='spr'></a>
#####   __Upper bound for a call spread__

$C(K_1) - C(K_2) <= e^{-rT}(K_2 - K_1)$, for $K_1 < K_2$

<br>

<a id='conv'></a>
#####   __Convexity__

Call price is a convex function of the exercise price $K$, i.e., 
 
$C(K_2) \le \frac{K_3 - K_2}{K_3 - K_1}C(K_1) + \frac{K_2 - K_1}{K_3 - K_1}C(K_3)$, for $K_1 < K_2 < K_3$

__*American Calls* (No Dividends)__ 

Suppose you are holding an ITM American call.

__What should you do: hold to expiration or exercise now?__ 

Reasons for not exercising a call early (no dividends): 

* No income from the stock is sacrificed

* Strike price is paid later (buy the asset at the strike) ñ time value of money advantage tells you to hold the option rather than exercise early

* Insurance premium of the option retained

__If you do not want to hold the call any longer, just sell it and receive its market
value, but do not exercise.__ 

1. Suppose you do exercise at some time $t^*$ before maturity.

   * __At time $t^*$__: <br> You get the stock worth $S(t^*)$ and pay the strike price $K$. The value: <br>

     $S_{t^*} - K$

   * __At time T__:<br> Your position is worth:

      $S_T - e^{r \tau}K$, where $\tau = T-t^*$

2. Alternatively, suppose you decide to hold to expiration. 

   * Then at time $T$ your payoff is:
   
     $(S_T - K)^{+}$
     
     
Since $(S(T) - K,0)^{+} \ge S(T) - e^{r \tau}K$ <br>
i.e., the call option payoff dominates the payoff of the alternative strategy of
exercising at $t^*$ , you should never exercise a call early

Price of American Call = Price of European Call

If you do exercise a call early, you lose:

* Interest on the amount $K$ you will have to pay until maturity $T$
* Insurance premium of the ITM call

<br>

##### __*American Puts* (No Dividends)__ 
Even in the absence of dividends it may be optimal to exercise an ITM put before
maturity. This happens when the stock falls low enough that the benefit received from the
likelihood that it falls even more is less than the interest that can be earned on the cash
received from immediate exercise. 

1. Suppose you do exercise at some time $t^*$ before maturity.

   * __At time $t^*$__: <br> You short the stock worth $S(t^*)$ and get the strike price $K$. The value: <br>

     $K - S(t^*)$      

   * __At time T__:<br> Your position is worth:

      $e^{r \tau}K - S(T)$       (1)

2. On the other hand, if we hold on to the option, we get the payoff:

  $(K - S_T)^{+}$             (2)

If $S(T) < e^{r \tau}K$: (1) is greater.
  
Otherwise, (2) is greater


* Time value of money tells us we should exercise and invest the sale proceeds to earn the interest

* But if we exercise we sacrifice the insurance value left in the put. 

Thus we exercise only if interest on the strike $K$ is greater than the remaining insurance value. 

<br>

##### __Early Exercise Summary:__

* Calls (no dividends): never optimal to exercise early

* Calls (cash dividends): may be optimal to exercise ITM calls just prior to an exdividend date.

* Calls (dividend yield): may be optimal to exercise ITM calls early

* Put (no dividends): maybe optimal to exercise ITM puts

* Puts (dividends): dividends increase the value of the put. If dividends are highenough, it may not be optimal to exercise in a situation where it would be optimal to exercise without dividends.

### __Options Positions and Trading Strategies__

#### __Applications of Option Strategies:__

1. __Insurance__ <br/>
    Options are purchased to protect against adverse changes in the value of   underlying asset, while retaining profit opportunity.

2. __Income (Yield) Enhancement__<br/>
   Options are sold to provide income in order to increase returns or lower costs.

3. __Trading/Speculative Positions__ <br/>
   Options are purchased and sold to take positions to profit on expected market or volatility
   moves.
   
4. __Trading/Arbitrage Positions__<br/>
   To capitalize on mispricing of options vs. the underlying asset or options vs. options.
